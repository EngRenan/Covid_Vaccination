Analise e coleta de dados provenientes do Kaggle a respeito de estatisticas das vacinas contra a pandemia de COVID 

```python
# Importando bibliotecas
import pandas as pd
import os
```

```python
# Importando dados do arquivo
Diretorio = '/content/'
arquivo = 'country_vaccinations.csv'
dados = os.path.join(Diretorio, arquivo) # Busca caminho do arquivo com suas conversões
```

```python
# Verifica o caminho do diretorio que o arquivo esta salvo
dados
```

Agora sera criado um data frame no pandas e ler o arquivo na pasta de origem. Para abrir o arquivo vamos utilizar o comando read_csv e utilizar o separador de ',' de acordo com o dado fornecido e tambem utilizar o encoding informando a codificação do arquivo

```python
dataframe = pd.read_csv(dados, sep=',', encoding='latin-1')
```

```python
# Exibindo os primeiros registros da tabela para verificação
dataframe.head(5)
```

# Analisando os dados

Para realizar a analise vamos responder algumas perguntas: 

1. Quais sao os paises com maior numero de pessoas imunizadas? 
2. Qual vacina é a mais utilizada no mundo?
3. Qual país possui o maior numero de vacinados por dia? 
4. Qual a media de vacinação por país? 

```python
# Verificando dados referente ao Brasil por dia (teste)
dataframe.query('country == "Brazil" ').sort_values('date', ascending=False)[:5]
```

```python
# Verificando o numero de paises distintos na base de dados
dataframe['country'].nunique()
```

Observado que temos registro de 223 paises no dataset

# 1. Qual o país com maior numero de pessoas imunizadas? 

Selecionaremos o ultimo registro (soma total) de cada país realizando um agrupamento do país e selecionando a ultima data de atualização que foi realizada

```python
dataframe.groupby('country')['date'].max().sort_values(ascending=False)
```

Visualizado que existem paises qie nao possuem a atualização da ultima data da coleta. Dessa forma iremos utilizar a varivavel total_vaccinations no qual contem o total de vacinações naquele país. Podemos usar como parametro para buscar a ultima data de coleta. Dessa forma é necessario agrupar por país e pegar o valor máximo obtido nessa variável.

```python
dataframe.groupby('country')['total_vaccinations'].max().sort_values(ascending=False)
```

```python
# Transformando em serie Pandas
imunizados_pais = dataframe.groupby('country')['total_vaccinations'].max().sort_values(ascending=False)
type(imunizados_pais)
```

```python
# Transformando em Dataframe novamente e resetar o index
imunizados = imunizados_pais.to_frame().reset_index()
imunizados
```

```python
# Criando lista dos top 10 paises 
imunizados.sort_values('total_vaccinations',ascending=False).head(10)
```

# 2. Qual a vacina mais utilizada no mundo?

Iremos criar um dataframe contendo as vacinas utilizadas por país. Com isso teremos uma relação com todas as vacinas por país em uma determinada data. 

```python
vacinas_por_pais = dataframe[['country', 'date', 'vaccines']]
vacinas_por_pais
```

```python
# Agrapando as vacinas e contando o país apenas uma vez. Se no dataset tiver um país utilizando duas vacinas, será contabilizado duas vezes. 
vacinas_por_pais.groupby('vaccines')['country'].nunique().sort_values(ascending=False)
```

```python
vacinas_pais = dataframe.groupby('country')['vaccines'].unique()
vacinas_pais = vacinas_pais.to_frame().reset_index()
```

Observado que Oxford/AstraZeneca é a vacina mais utilizada pelos paises  

# 3. Qual país possui o maior numero de vacinados por dia?

```python
# Iremos avaliar a variavel daily_vaccinations
dataframe.groupby('country')['daily_vaccinations'].max().sort_values(ascending=False)
```

Analisado e visto que China, India, Bangladesh e United States tem o maior numero de vacinados por dia. 

# Qual a media de vacinação por país?

```python
media_pais = dataframe.groupby('country')['daily_vaccinations'].mean().sort_values(ascending=False)
media_pais = media_pais.to_frame().reset_index()
media_pais
```

Paises que tem a maior media de vacinação sao: China, India, USA, Brazil e Indonesia

# Criando base de dados somente com os dados tratados anteriormente

```python
# Renomeando colunas 
media_pais = media_pais.rename(columns={'country': 'pais', 'daily_vaccinations': 'media_vacinacao'})
media_pais
```

```python
vacinas_pais = vacinas_pais.rename(columns={'country': 'pais', 'vaccines': 'vacina_utilizada'})
vacinas_pais[:5]
```

```python
imunizados = imunizados.rename(columns={'country': 'pais', 'total_vaccinations': 'Imunizados'})
imunizados[:5]
```

# Criando novo dataset

Para criar um novo dataset sera necessário utilizar a função merge no qual vai comparar os campos das tabelas que são chaves iguais, como o nome do país e realizar uma junção entre as tabelas, lendo todos os dados tratados em uma unica tabela. 

```python
# Realizando merge
dados_tratados = pd.merge(imunizados, media_pais, how='inner')

dados_tratados = pd.merge(dados_tratados, vacinas_pais)
dados_tratados[:10]
```

# Exportando dados tratados

```python
dados_tratados.to_csv('estatistica_pais.csv', index=True, sep=';')
```

```python
dados_tratados.to_json('estatistica.json', orient='records')
```
